```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
greedy = pd.read_csv('greedy_self_play_fixed.csv')
maxn = pd.read_csv('maxn_self_play_fixed.csv')
paranoid = pd.read_csv('paranoid_self_play_fixed.csv')
```

```python
def depth_graph(data, title):

    labels = data.id[:50]
    p1distance = data.p1distance[:50]
    p2distance = data.p2distance[:50]
    p3distance = data.p3distance[:50]

    x = np.arange(len(labels))  # the label locations
    width = 0.2  # the width of the bars

    fig, ax = plt.subplots(figsize = [15,5])
    rects1 = ax.bar(x - width * 3/2, p1distance, width, label="depth1")
    rects2 = ax.bar(x, p3distance, width, label="depth2")
    rects3 = ax.bar(x + width * 3/2, p2distance, width, label = "depth3")

    # Add some text for labels, title and custom x-axis tick labels.
    ax.set_ylabel('Distance from wining state')
    ax.set_xlabel('Experiment number')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)

    # declutter the x-axis labels
    #every_nth = 2
    #for n, label in enumerate(ax.xaxis.get_ticklabels()):
        #if n % every_nth != 0:
            #label.set_visible(False)

    ax.legend(loc = 9)
    fig.tight_layout()
    
    return plt.show()
```

```python
depth_graph(maxn, 'maxN depths 1-3')
```

```python
maxnD1avg = maxn.p1distance.sum() / 50
maxnD2avg = maxn.p3distance.sum() / 50
maxnD3avg = maxn.p2distance.sum() / 50

maxnD1movesavg = maxn.p1moves.sum() / 50
maxnD2movesavg = maxn.p3moves.sum() / 50
maxnD3movesavg = maxn.p2moves.sum() / 50

maxnD1thinkingavg = maxn.p1thinkingtime.sum() / 50
maxnD2thinkingavg = maxn.p3thinkingtime.sum() / 50
maxnD3thinkingavg = maxn.p2thinkingtime.sum() / 50

winrate = maxn['winner'].value_counts(normalize=True)

print ("depth1 avg: " , maxnD1avg, "avg moves: " , maxnD1movesavg, " avg thinking: " , maxnD1thinkingavg)
print ("depth2 avg: " , maxnD2avg, "avg moves: " , maxnD2movesavg, " avg thinking: " , maxnD2thinkingavg)
print ("depth3 avg: " , maxnD3avg, "avg moves: " , maxnD3movesavg, " avg thinking: " , maxnD3thinkingavg)

print()
print(winrate)
```

```python
depth_graph(paranoid, 'paranoid-depths 1-3')
```
